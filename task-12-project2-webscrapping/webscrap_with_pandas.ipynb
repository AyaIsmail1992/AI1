```python
import requests 
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

```python
website = requests.get("https://www.imdb.com/chart/top/").text
soup = BeautifulSoup(website, 'lxml') # parsing (understand txt string as html tags, id, class)
print(soup.prettify())
```

```python

#csvfile = open ('scrape-imdb.csv', 'w', encoding='utf-8', newline='') # 'w' = write into the file
#writer = csv.writer(csvfile) # import csv
#writer.writerow(['Movie Rank', 'Movie Title', 'Year','Rate'])#

movies=soup.find('tbody',class_='lister-list')
#print(movies)
all_movies=movies.find_all('tr')
#print(all_movies)
movie_rank=[]
movie_title=[]
movie_year=[]
movie_rate=[]

for mov in all_movies:
    #print("mov: ", mov)
    movie_data=[]
    
    
    title_rank_year=mov.find('td',class_='titleColumn')#=mov.td.a
    #print('title_rank_year: ',title_rank_year)
    
    mov_data=title_rank_year.text #rank, title, year
    mov_data=mov_data.strip().split('\n')   #split(' ',1)
    #print('mov_data: ',mov_data)
    
    mov_rank=mov_data[0].split('.')[0]
    
    movie_data.append(mov_rank)
    movie_rank.append(mov_rank)
    
    #print("mov_rank: ",mov_rank)#1
    mov_title=mov_data[1].strip()
    movie_data.append(mov_title)
    movie_title.append(mov_title)
    
    #print('mov_title: ',mov_title)
    mov_year=mov_data[-1].strip().replace("(","").replace(")","")
    print('mov_year: ',mov_year)
    movie_data.append(mov_year)
    movie_year.append(mov_year)
    
    rate=mov.find('td',class_='ratingColumn imdbRating')#=mov.td.a
    #print('rate: ',rate)
    mov_rate=rate.text
    movie_data.append(mov_rate.replace("\n", ""))
    print('mov_rate: ',mov_rate.replace("\n", ""))
    movie_rate.append(mov_rate.replace("\n", ""))
   
    print('movie_data: ',movie_data)
    
    #writer.writerow(movie_data) 
    """df.append({"Movie Rank":mov_rank,
               "Movie Title": mov_title,
               "Year":mov_year,
               "Rate":mov_rate},ignore_index=True)
    """
    
#csvfile.close()
```

```python
data={"Movie Rank":movie_rank,
               "Movie Title": movie_title,
               "Year":movie_year,
               "Rate":movie_rate}
df=pd.DataFrame(data)
print(df)
```
